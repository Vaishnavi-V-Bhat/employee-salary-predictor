```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(n_estimators=100),
    "Gradient Boosting": GradientBoostingClassifier(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "CatBoost": CatBoostClassifier(verbose=0)
}

results ={}
for name, model in models.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    acc= accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"{name}: {acc:.4f}")
    print(classification_report(y_test, y_pred))
```

```python
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split, RandomizedSearchCV
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# # Load data
# data = pd.read_csv('/workspaces/Employee_Salary_prediction/adult 3.csv')

# # Preprocessing: drop rows with missing values (if any)
# data = data.replace('?', pd.NA).dropna()

# # Encode categorical columns using pandas get_dummies
# X = data.drop('income', axis=1)
# y = data['income']

# # Map target to numeric salary values
# y = y.map({'<=50K': 35000, '>50K': 75000})

# One-hot encode categorical features
X_encoded = pd.get_dummies(X)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Define CatBoostRegressor
cat_reg = CatBoostRegressor(random_state=42, verbose=0, eval_metric='R2')

# Define parameter distribution for RandomizedSearchCV
param_dist = {
	'iterations': [100, 250, 500, 1000],
	'learning_rate': [0.01, 0.05, 0.1, 0.2],
	'depth': [4, 6, 8, 10],
	'l2_leaf_reg': [1, 3, 5, 7, 9],
	'border_count': [32, 64, 128, 254],
	'bagging_temperature': [0.5, 0.8, 1],
	'random_strength': [0.5, 1, 1.5]
}

# RandomizedSearchCV
random_search = RandomizedSearchCV(
	estimator=cat_reg,
	param_distributions=param_dist,
	n_iter=20,
	cv=3,
	scoring='r2',
	random_state=42,
	n_jobs=-1
)

# Fit the model
random_search.fit(X_train, y_train)

# Best estimator
best_cat_reg = random_search.best_estimator_

# Predict and evaluate

y_pred = best_cat_reg.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Best Parameters:", random_search.best_params_)
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)
```
